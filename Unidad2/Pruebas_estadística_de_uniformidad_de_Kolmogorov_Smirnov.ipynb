```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def generar_numeros_aleatorios(n=100):
    return np.random.uniform(0, 1, n)

def prueba_kolmogorov_smirnov(datos):
    estadistico, p_valor = stats.kstest(datos, 'uniform', args=(0, 1))
    return estadistico, p_valor

def graficar_histograma(datos):
    plt.hist(datos, bins=10, edgecolor='black', alpha=0.7, density=True)
    plt.axhline(y=1, color='r', linestyle='--', label='Distribución uniforme ideal')
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')
    plt.title('Histograma de números aleatorios')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    numeros = generar_numeros_aleatorios()
    estadistico, p_valor = prueba_kolmogorov_smirnov(numeros)

    print(f'Estadístico de Kolmogorov-Smirnov: {estadistico}')
    print(f'Valor p: {p_valor}')

    if p_valor > 0.05:
        print("No se rechaza la hipótesis nula: los datos parecen provenir de una distribución uniforme.")
    else:
        print("Se rechaza la hipótesis nula: los datos no siguen una distribución uniforme.")

    graficar_histograma(numeros)
```
