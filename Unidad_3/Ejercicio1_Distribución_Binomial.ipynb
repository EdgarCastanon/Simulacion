# Simulación de Pedidos Satisfechos con Distribución Binomial  

## Introducción  
Este programa utiliza la distribución binomial para simular la cantidad de pedidos que un distribuidor de artículos electrónicos logra satisfacer diariamente durante una semana.  

## Objetivo  
El objetivo es modelar el número de pedidos satisfechos en función de la probabilidad de éxito (70%) y el número total de pedidos diarios (20).  

## Pasos del Código  
1. **Definir Parámetros**: Se establecen los valores para la cantidad de pedidos diarios, la probabilidad de éxito y la duración de la simulación (7 días).  
2. **Simulación con Distribución Binomial**: Se generan valores aleatorios para representar la cantidad de pedidos satisfechos cada día.  
3. **Cálculo de Estadísticas**: Se obtienen la media, la desviación estándar, el máximo y el mínimo de pedidos satisfechos.  
4. **Visualización de Resultados**:  
   - **Gráfico de barras**: Representa el número de pedidos satisfechos cada día.  
   - **Histograma**: Muestra la distribución de la cantidad de pedidos satisfechos en la semana.  

## Conclusión  
Este modelo permite al distribuidor anticipar la variabilidad en el cumplimiento de pedidos y optimizar su gestión de inventario.  


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parámetros de la distribución binomial
n_pedidos = 20  # Número de pedidos por día
p_exito = 0.7  # Probabilidad de éxito (satisfacer un pedido)
dias = 7  # Días en la semana

# Simulación de la cantidad de pedidos satisfechos por día en una semana
pedidos_satisfechos = np.random.binomial(n_pedidos, p_exito, dias)

# Cálculo de estadísticas
media = np.mean(pedidos_satisfechos)
desviacion = np.std(pedidos_satisfechos)
max_satisfechos = np.max(pedidos_satisfechos)
min_satisfechos = np.min(pedidos_satisfechos)

# Mostrar los resultados
print("Pedidos satisfechos cada día de la semana:", pedidos_satisfechos)
print(f"Media de pedidos satisfechos: {media:.2f}")
print(f"Desviación estándar: {desviacion:.2f}")
print(f"Máximo de pedidos satisfechos en un día: {max_satisfechos}")
print(f"Mínimo de pedidos satisfechos en un día: {min_satisfechos}")

# Configurar estilo del gráfico
sns.set_style("whitegrid")
plt.figure(figsize=(10, 5))
plt.bar(range(1, dias + 1), pedidos_satisfechos, color='skyblue', edgecolor='black')
plt.axhline(media, color='red', linestyle='dashed', label=f'Media: {media:.2f}')
plt.xlabel('Día de la semana')
plt.ylabel('Pedidos satisfechos')
plt.title('Simulación de Pedidos Satisfechos por Día')
plt.xticks(range(1, dias + 1))
plt.legend()
plt.show()

# Histograma de distribución de pedidos satisfechos
plt.figure(figsize=(8, 5))
sns.histplot(pedidos_satisfechos, bins=range(min_satisfechos, max_satisfechos + 2), kde=True, color='blue')
plt.xlabel('Pedidos satisfechos')
plt.ylabel('Frecuencia')
plt.title('Distribución de Pedidos Satisfechos en la Semana')
plt.show()
```
